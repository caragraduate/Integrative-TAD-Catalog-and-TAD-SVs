## TAD-SV-eqtl

```python
sig_del_eqtl = pd.read_csv('../data/sig_del_eqtl.bed', sep = '\t')
sig_ins_eqtl = pd.read_csv('../data/sig_ins_eqtl.bed', sep = '\t')
sig_del_eqtl_new = sig_del_eqtl[['gene', 'INFO_ID', 'NA18534','NA18939','NA19036','NA19240','NA19650','NA19983','NA20509','NA20847','HG00096','HG00171','HG00514','HG00733','HG00864','HG01114','HG01505','HG01573','HG01596','HG02011','HG02018','HG02492','HG02587','HG03009','HG03065','HG03371','HG03683','HG03732']]
sig_ins_eqtl_new = sig_ins_eqtl[['gene', 'INFO_ID', 'NA18534','NA18939','NA19036','NA19240','NA19650','NA19983','NA20509','NA20847','HG00096','HG00171','HG00514','HG00733','HG00864','HG01114','HG01505','HG01573','HG01596','HG02011','HG02018','HG02492','HG02587','HG03009','HG03065','HG03371','HG03683','HG03732']]
sig_DEL_boundary_genotype = DEL_boundary_genotype[DEL_boundary_genotype['INFO_ID'].isin(sig_del_eqtl['INFO_ID'])]
sig_DEL_boundary_genotype_new = sig_DEL_boundary_genotype[['INFO_ID', 'NA18534','NA18939','NA19036','NA19240','NA19650','NA19983','NA20509','NA20847','HG00096','HG00171','HG00514','HG00733','HG00864','HG01114','HG01505','HG01573','HG01596','HG02011','HG02018','HG02492','HG02587','HG03009','HG03065','HG03371','HG03683','HG03732']]
sig_INS_boundary_genotype = INS_boundary_genotype[INS_boundary_genotype['INFO_ID'].isin(sig_ins_eqtl['INFO_ID'])]
sig_INS_boundary_genotype_new = sig_INS_boundary_genotype[['INFO_ID', 'NA18534','NA18939','NA19036','NA19240','NA19650','NA19983','NA20509','NA20847','HG00096','HG00171','HG00514','HG00733','HG00864','HG01114','HG01505','HG01573','HG01596','HG02011','HG02018','HG02492','HG02587','HG03009','HG03065','HG03371','HG03683','HG03732']]
sig_DEL_boundary_3 = all_DEL_boundary_3_sig[all_DEL_boundary_3_sig['INFO_ID'].isin(sig_del_eqtl['INFO_ID'])]
sig_DEL_boundary_3_new = sig_DEL_boundary_3[['INFO_ID', 'GM18534','GM18939','GM19036','GM19240','GM19650','GM19833','GM20509','GM20847','HG00096','HG00171','HG00514','HG00733','HG00864','HG01114','HG01505','HG01573','HG01596','HG02011','HG02018','HG02492','HG02587','HG03009','HG03065','HG03371','HG03683','HG03732']]
sig_INS_boundary_3 = all_INS_boundary_3_sig[all_INS_boundary_3_sig['INFO_ID'].isin(sig_ins_eqtl['INFO_ID'])]
sig_INS_boundary_3_new = sig_INS_boundary_3[['INFO_ID', 'GM18534','GM18939','GM19036','GM19240','GM19650','GM19833','GM20509','GM20847','HG00096','HG00171','HG00514','HG00733','HG00864','HG01114','HG01505','HG01573','HG01596','HG02011','HG02018','HG02492','HG02587','HG03009','HG03065','HG03371','HG03683','HG03732']]
```

```python
sig_del_eqtl_new = sig_del_eqtl_new.copy()
for index, row in sig_del_eqtl_new.iterrows():
    gene = row['gene']
    info_id = row['INFO_ID']
    
    # Extract rows from other datasets using INFO_ID
    boundary_row = sig_DEL_boundary_3_new[sig_DEL_boundary_3_new['INFO_ID'] == info_id]
    genotype_row = sig_DEL_boundary_genotype_new[sig_DEL_boundary_genotype_new['INFO_ID'] == info_id]
     
    # Check if the rows for the current INFO_ID are available in both datasets
    if boundary_row.empty or genotype_row.empty:
        continue
    
    # Prepare the data for analysis
    Gene_expression = row.drop(labels=['INFO_ID', 'gene']).values.flatten()
    Bound_strength = boundary_row.drop(columns=['INFO_ID']).values.flatten()
    SV = genotype_row.drop(columns=['INFO_ID']).values.flatten()
    
    # Create a DataFrame for the current gene and INFO_ID
    df = pd.DataFrame({
        'Gene_expression': Gene_expression,
        'Bound_strength': Bound_strength,
        'SV': SV
    })
    
    # Convert entire DataFrame to numeric, setting non-numeric values to NaN
    df = df.apply(pd.to_numeric, errors='coerce')
    
    # Drop rows with NaN values (if any) due to type conversion
    df = df.dropna()
    
    # Ensure that there are no missing values
    if df.empty:
        continue
    
    # Set a random seed for reproducibility
    np.random.seed(710)

    # The mediator model aims to examine how the independent variable (SV) affects the mediator variable (boundary strength).
    mediator_formula = 'Bound_strength ~ SV'
    # The outcome model examines how the independent variable (SV) and the mediator variable (boundary strength) together affect the dependent variable (gene expression).
    outcome_formula = 'Gene_expression ~ SV + Bound_strength'

    # Fit the mediator and outcome models using formulas
    mediator_model = sm.OLS.from_formula(mediator_formula, df)
    outcome_model = sm.OLS.from_formula(outcome_formula, df)
    
    # Perform mediation analysis
    mediation_analysis = Mediation(outcome_model, mediator_model, 'SV', 'Bound_strength').fit()
    
    # Extract p-values from the mediation analysis summary
    mediation_summary = mediation_analysis.summary()
    pval_ACME = mediation_summary.loc['ACME (average)', 'P-value']
    pval_ADE = mediation_summary.loc['ADE (average)', 'P-value']
    pval_Total = mediation_summary.loc['Total effect', 'P-value']
    
    # Store the p-values in the sig_del_eqtl_new DataFrame
    sig_del_eqtl_new.at[index, 'pval_ACME'] = pval_ACME
    sig_del_eqtl_new.at[index, 'pval_ADE'] = pval_ADE
    sig_del_eqtl_new.at[index, 'pval_Total'] = pval_Total
```

```python
sig_del_eqtl_new.to_csv('../data/all_del_eqtl_mediation.bed', index = False, sep='\t', header=True)
```
